# Copias Leibles

Los archivos descargados de la página no son operables\
Es aquí donde creamos copias reparadas

```python
# Damos a una variable los nombres de cada archivo ubicado en la carpeta 'Iniciadas_Peru_Mundo'
xlsb_x = os.listdir()

"""
 Crearemos copias de los 'xlsb' como xlsx.
"""
x = 1
for i in xlsb_x:
    try:
        o = win32com.client.Dispatch("Excel.Application")
        o.Visible = False

        filename = os.getcwd() + '/' + i
        output = os.getcwd() + '/' + f'reparado_{x}.xlsx'

        wb = o.Workbooks.Open(filename)
        wb.ActiveSheet.SaveAs(output,51)
        x += 1
    except:
        pass
```

```python
"""
Agrupamos los Excel a usar
"""
xlsb_x = os.listdir()
# Esta linea separa archivos que contengan el siguiente string
xlsb_x = [f for f in xlsb_x if f.startswith('rep')]
```

# Union y Limpieza
Unimos los archivos\
Borramos 9 primeras filas\
Borramos columnas con datos innecesarios\
Borramos filas y columnas vacías 'NAN'

```python
"""
Unimos los Excel
"""
# Borramos las primeras 9 lineas.
df = pd.read_excel(xlsb_x[0],  skiprows=9)

for i in xlsb_x[1:-1]:
    
    df = df.append(pd.read_excel(i,  skiprows=9))
frame  = df.drop(columns = ['Fuentes y referencias','Unnamed: 18'])
# Confirmamos el índice defectuoso por el Merch de los archivos y la existencia de valores 'NAN'
frame
```

```python
"""
Limpieza
"""
# Borramos filas vacías
fram = frame.dropna(how = 'all')
# Borramos columnas vacías
fram = fram.dropna(axis = 1, how = 'all')
# Ordenamos por fecha y por pais
fram = fram.sort_values(['En preparación', 'Socio afectado'])
# Receteamos el indice
fram = fram.reset_index()
# Se ha creado una columna 'index' la eliminamos asi:
fram.drop(['index'], axis=1, inplace=True)
# Verificamos que todo está en orden.
fram
```

```python
# Creamos Excel XLSX.
# Este es el resultado de todas las iteraciones y limpieza.
fram.to_excel('Peru_Mundo.xlsx')
```
